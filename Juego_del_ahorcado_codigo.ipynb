````python
# FUNCIONES

# FUNCION JUEGO NUCLEAR

def basic_game(word, hint, gamer):
    game_win = 0
    game_points = 0
    word_hidden_list = [" _ "] * len(word)  # Palabra a adivinar oculta
    word_hidden = "".join(word_hidden_list)
    used_letters = set()     # Set letras usadas en partida
    failed_letters = 0
    lives = 10            # Vidas iniciales con las que empezamos las partidas 
    print(f"\n\n¡Empecemos {gamer}! Esta es la palabra a adivinar {word_hidden}\n")
    print(f"{ahorcado_pics[lives]}\n")
    
    try:   
        while lives>0 and " _ " in word_hidden_list:
            letter = input("Introduce tu letra. Puedes indicar la palabra 'Pista' para recibir una ayudita especial").upper()
            if letter == "PISTA":
                print(f"La pista es: '{hint}'\n")
            elif letter.isalpha() and len(letter) == 1:
                    if letter in used_letters:
                        print("¡Esa letra ya la escribiste! \n¿Estás aplicando recursividad para fallar la misma letra en bucle?\n")
                        print(f"Estas son las letras utilizadas hasta el momento: {",".join(sorted(used_letters))}\n")  
                    elif letter in word:
                        for letter_word in range(len(word)):
                            if word[letter_word] == letter:
                                word_hidden_list[letter_word] = letter
                                word_hidden = "".join(word_hidden_list)
                                used_letters.add(letter)
                        print(f"¡Ole, ole y ole! ¡Acertaste! \nLa letra {letter} está en la palabra. \n-------------------\n\n  {word_hidden}  \n\n-------------------\n")
                    else: 
                        if lives > 1:
                            lives -= 1
                            print(f"{ahorcado_pics[lives]}\n")
                            print(f"{dicc_motivation[lives]}\n")
                            print(f"Vaya, la letra {letter} no está en la palabra {word_hidden}.\nTe quedan {lives} vidas.\n")
                            used_letters.add(letter)
                            failed_letters += 1
                            print(f"Estas son las letras utilizadas hasta el momento: {",".join(sorted(used_letters))}\n")
                        elif lives == 1:
                            print(f"G A M E   O V E R, ¡seguro que con las funciones se te da mejor!\nLa palabra era {word}")
                            lives -= 1
                            print(ahorcado_pics[lives])
                            print(dicc_motivation[lives])
                            used_letters.add(letter)
                            failed_letters += 1
                            game_points += (failed_letters*5)
            else: 
                print("Recordemos las normas: solo UNA y debe ser LETRA\n")
        if " _ " not in word_hidden_list:
            print(f"¡¡E N H O R A B U E N A!! \n{gamer}, descubriste la palabra {word} con {10-lives} fallos, no esperaba menos ^_^ \nYa puedes empezar con el siguiente módulo de Adalab, estás preparada\n")
            game_win +=1
            game_points += (failed_letters*5)
        return game_win, game_points
    except TypeError:
        print("El carácter introducido da error\n")
    except ValueError:
        print("El carácter introducido da error\n")

#FUNCIÓN PARA VERIFICAR PALABRAS Y PISTAS

def min_letters_word(intro):
    while len(intro) < 2 or not intro.isalpha():
            intro = input(f"Procura que la palabra al menos tenga dos carácteres y sean letras").strip().upper()
    return intro

def min_letters_hint(intro):
    while len(intro) < 2 or not intro.isalpha():
            intro = input(f"Para la pista es necesario que sea una palabra de, al menos, dos letras").strip().upper()
    return intro

````
